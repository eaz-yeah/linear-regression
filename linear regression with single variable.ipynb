```python
import numpy as np
```

```python
#so we loaded the csv file into the data which act as numpy array
data=np.loadtxt("/content/drive/MyDrive/Colab Notebooks/linear regression with single varible/single_variable.csv", delimiter=",")
data.shape
```

```python
data[0:5, : ]
```

```python
#the first columns will act as input whereas the second columns will act as the output
input=data[: , 0]
output=data[: ,1]
input.shape, output.shape
```

```python
input=input.reshape(-1,1)
output=output.reshape(-1,1)
```

```python
input.shape, output.shape
```

```python
#now i will implement model_selction from sklearn so that i can split my data in the form of x_train, x_test, y_train, y_test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(input, output, random_state=4, test_size=0.2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#now we will implement the linear regession model
from sklearn.linear_model import LinearRegression
algo=LinearRegression()
algo.fit(x_train, y_train)
```

```python
algo.score(x_train, y_train), algo.score(x_test, y_test)
```

```python
#now lets compute y_train_pred, y_test_pred using the linear regression model
y_train_pred=algo.predict(x_train)
y_test_pred=algo.predict(x_test)
```

```python
algo.coef_, algo.intercept_
```

```python
m=algo.coef_[0,0]
c=algo.intercept_[0]
print("the slope and intercept of the line formed through linear regression:-", m ,  ", ", c)
```

```python
#lets plot the graph of y_pred_train and y_pred_test
import matplotlib.pyplot as plt
plt.scatter(y_train,y_train_pred)

x_value=np.arange(0,25,0.5)
y_value=x_value  #the y_value represents the regression line
plt.plot(x_value, y_value,color="red")
plt.xlabel("y_train")
plt.ylabel("y_train_pred")
plt.title("the comparision between y_train and y_train_pred",color="green")
```

```python
#lets plot the graph of y_test_pred and y_test

plt.scatter(y_test,y_test_pred)

x_value=np.arange(0,25,0.5)
y_value=x_value  #the y_value represents the regression line
plt.plot(x_value, y_value,color="red")
plt.xlabel("y_test")
plt.ylabel("y_test_pred")
plt.title("the comparision between y_test and y_test_pred",color="green")
```

```python
#lets form the regression line between x_train and y_train
plt.scatter(x_train,y_train)

x_value=np.arange(0,25,0.5)
y_value=x_value * m + c  #the y_value represents the regression line
plt.plot(x_value, y_value,color="red")
plt.xlabel("x_train")
plt.ylabel("y_train")
plt.title("the comparision between x_train and y_train",color="green")
```
